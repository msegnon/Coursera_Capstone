## Methodology <a name="methodology"></a>

### Cluster the Queens and Scarborough Borough using K-Means

```python
# set number of clusters
kclusters = 8

toronto_grouped_clustering = scarborough_grouped.drop('Neighbourhood', 1)
queens_grouped_clustering = queens_grouped.drop('Neighbourhood', 1)

# run k-means clustering
queens_cluster_labels = utils.return_labels ( n_clusters= kclusters, X =queens_grouped_clustering, random_state = 4  )
scarborough_cluster_labels = utils.return_labels ( n_clusters= kclusters, X =toronto_grouped_clustering, random_state = 0  )
```

```python
queens_cluster_labels[:10]
```

```python
scarborough_cluster_labels[:10]
```

Lets create a new dataframe that includes the cluster as well as the top ten venues for each neighborhood

```python
#Note that the neighborhood Upper Rouge does not have any venues, so I will drop from dataset
scarborough_data.drop(scarborough_data[scarborough_data.Neighbourhood == 'Upper Rouge'].index, inplace = True)

scarborough_merged = scarborough_data

# add clustering labels
scarborough_merged['Cluster Labels'] = scarborough_cluster_labels

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
scarborough_merged = scarborough_merged.join(neighborhoods_venues_sorted_scarborough.set_index('Neighbourhood'), on='Neighbourhood')

scarborough_merged.head() # check the last columns!
```

```python
queens_merged = queens_data

# add clustering labels
queens_merged['Cluster Labels'] = queens_cluster_labels

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
queens_merged = queens_merged.join(neighborhoods_venues_sorted_queens.set_index('Neighbourhood'), on='Neighbourhood')

queens_merged.head() # check the last columns!
```

```python
# create map
map_clusters_scarboroguh = utils.create_map_clusters(kclusters=kclusters, latitude=location_scarborough.latitude, longitude=location_scarborough.longitude, data_merged = scarborough_merged, zoom_start=10 )
map_clusters_queens = utils.create_map_clusters(kclusters=kclusters, latitude=location_queens.latitude, longitude=location_queens.longitude, data_merged = queens_merged, zoom_start=10 )
```
